```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_file = "owid-covid-data.csv"
raw_data_info = pd.read_csv(data_file)
raw_data_info.head()
```

```python
grouped_countries_df = raw_data_info.groupby("location")
```

```python
age_65_older = grouped_countries_df[['aged_65_older','cvd_death_rate']].max()
age_65_older=age_65_older.dropna()
age_65_older
```

```python
plt.scatter(age_65_older['aged_65_older'], age_65_older['cvd_death_rate'])
plt.title("age_65_older vs. cvd_death_rate")
plt.xlabel("age_65_older")
plt.ylabel("cvd_death_rate")
plt.show()
```

```python
age_65_older = grouped_countries_df[['aged_65_older','total_deaths_per_million']].max()
age_65_older=age_65_older.dropna()
age_65_older
```

```python
plt.scatter(age_65_older['aged_65_older'], age_65_older['total_deaths_per_million'])
plt.title("age_65_older vs. total_deaths_per_million")
plt.xlabel("age_65_older")
plt.ylabel("total_deaths_per_million")
plt.show()
```

```python
diabetes_prevalence = grouped_countries_df[['diabetes_prevalence','total_deaths_per_million']].max()
diabetes_prevalence=diabetes_prevalence.dropna()
diabetes_prevalence
```

```python
hospital_beds_per_thousand = grouped_countries_df[['hospital_beds_per_thousand','total_deaths_per_million']].max()
hospital_beds_per_thousand=hospital_beds_per_thousand.dropna()
hospital_beds_per_thousand
```
